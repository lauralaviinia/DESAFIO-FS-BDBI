##Importando bibliotecas

```python
#renomeando as bibliotecas para melhor praticidade

import pandas as pd
import matplotlib.pyplot as plt
```

##Carregando dados

```python
#acessando a biblioteca com o pd
#lendo os dados do arquivo baixado com o comando read

df = pd.read_csv('/content/dados_rh.csv')
df
```

##Agrupamento de Dados e Plotagem de Gráficos  

```python
#a.	Plote um gráfico de barras que exiba a quantidade de funcionários em cada setor

df2 = df['Department'].value_counts()

#value counts conta os valores únicos em uma série ou coluna

#estilizando o gráfico
#modelo, color, título, nomes dos eixos, rotação

df2.plot(kind='bar', color='#9c74c1')
plt.title('Quantidade de Funcionários por Setor')
plt.xlabel('Setor')
plt.ylabel('Quantidade de Funcionários')
plt.xticks(rotation=45)
```

```python
#b.	Plote um gráfico de linha que exiba a quantidade de funcionários para cada nível de escolaridade

df3 = df['Education'].value_counts().sort_index()

#sort index ordena um dataframe ou uma series com base nos seus índices em vez dos valores

#estilizando o gráfico

df3.plot(kind='line', color='#72429e')
plt.title('Quantidade de Funcionários por Nível de Escolaridade')
plt.xlabel('Nível de Escolaridade')
plt.ylabel('Quantidade de Funcionários')
```

```python
#c.	Plote um gráfico de torta/pizza que exiba a quantidade de funcionários que tem ou não algum atrito na empresa

df4 = df['Attrition'].value_counts()

#estilizando o gráfico
#autopct formata os valores percentuais dentro do gráfico

df4.plot(kind='pie', autopct='%1.1f%%', colors=['#190ea2', '#1c9ec8'])
plt.title('Proporção de Funcionários com ou sem Atrito')
plt.ylabel('')
```
